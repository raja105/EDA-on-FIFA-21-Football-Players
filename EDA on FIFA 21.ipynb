## Selecting particular number of columns to impute values

```python
fill_values=data.loc[:,['pace',
 'shooting',
 'passing',
 'dribbling',
 'defending',
 'physic']]

for i in fill_values.columns:
    data[i].fillna(data[i].mean(),inplace=True)
```

```python
data.isnull().sum()
```

## Plotting count of players on their heights

```python
plt.figure(figsize=(22,8))
ax=sns.countplot(x='height_cm',data=data)
ax.set_title(label="count of players on their heights",fontsize=20)
ax.set_xlabel(xlabel='Height',fontsize=16)
ax.set_ylabel(ylabel='Number of players',fontsize=16)
plt.show()
```

## Height vs Dribbling 

```python
plt.figure(figsize=(22,8))
plt.xlabel("Height",fontsize=20)
plt.ylabel("Dribbling",fontsize=20)
plt.title("Height vs Dribbling",fontsize=20)
sns.barplot(x='height_cm',y='dribbling',data=data.sort_values('height_cm',inplace=False))
plt.show()
```

```python
# We can see and analyse that with increase in height there is a decrease in the dribbling ability
```

## Weight vs Dribbling 

```python
plt.figure(figsize=(22,8))
plt.xlabel("Weight",fontsize=20)
plt.ylabel("Dribbling",fontsize=20)
plt.title("Weight vs Dribbling",fontsize=20)
sns.barplot(x='weight_kg',y='dribbling',data=data.sort_values('weight_kg',inplace=False))
plt.show()
```

```python
# not like increase in weight is affecting much
```

## Messi V/S Ronaldo

```python
skills=[
    'pace',
 'shooting',
 'passing',
 'dribbling',
 'defending',
 'physic',
 'attacking_crossing',
 'attacking_finishing',
 'attacking_heading_accuracy',
 'attacking_short_passing',
 'attacking_volleys',
 'skill_dribbling',
 'skill_curve',
 'skill_fk_accuracy',
 'skill_long_passing',
 'skill_ball_control',
 'movement_acceleration',
 'movement_sprint_speed',
 'movement_agility',
 'movement_reactions',
 'movement_balance',
 'power_shot_power',
 'power_jumping',
 'power_stamina',
 'power_strength',
 'power_long_shots',
 'mentality_aggression',
 'mentality_interceptions',
 'mentality_positioning',
 'mentality_vision',
 'mentality_penalties',
 'mentality_composure',
 'defending_marking',
 'defending_standing_tackle',
 'defending_sliding_tackle'
]
```

```python
messi=data.loc[data['short_name']=='L. Messi']
messi= pd.DataFrame(messi,columns=skills)

ronaldo=data.loc[data['short_name']=='Cristiano Ronaldo']
ronaldo= pd.DataFrame(ronaldo,columns=skills)
```

```python
messi
```

```python
ronaldo
```

```python
plt.figure(figsize=(30,10))
sns.pointplot(data=messi,color='red')
sns.pointplot(data=ronaldo,color='blue')
plt.title("Messi vs Ronaldo",fontsize=40)
plt.xticks(rotation=90)
plt.xlabel=("skills")
plt.ylabel("skills value",fontsize=20)
plt.grid()
```

## Age Distribution of players in Countries

```python
country_names=('Argentina','France','Germany','Belgium','Portugal','Poland','Brazil' )

countries= data.loc[data['nationality'].isin(country_names)& data['age']]

fig,ax=plt.subplots()
fig.set_size_inches(20,10)
ax=sns.boxenplot(x='nationality',y='age',data=countries)
plt.grid()
```

## Age Distribution of Players in clubs

```python
c=('FC Barcelona','Juventus','Real Madrid','Manchester City','Manchester United','Paris Saint-Germain' )

clubs= data.loc[data['club_name'].isin(c)& data['age']]

fig,ax=plt.subplots()
fig.set_size_inches(20,10)
ax=sns.boxenplot(x='club_name',y='age',data=clubs)
plt.grid()
```
